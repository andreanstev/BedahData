```python
#dfhappiness2020_2 = dataset happiness index yang sudah digabungkan dengan education score
dfhappiness2020_2 = pd.concat([dfhappiness2020_1,dfeducation], axis = 1)
```

```python
null_index = dfhappiness2020_2[dfhappiness2020_2['Regional indicator'].isnull() == True].index
```

```python
dfhappiness2020_2.drop(null_index, axis = 0, inplace = True) #Negara yang tidak masuk ke education ranking (null) diisi nol
```

```python
dfhappiness2020_2.fillna(0, inplace = True)
```

```python
dfhappiness2020_2.drop(['readingScore','mathScore','scienceScore','upperwhisker','lowerwhisker'], axis = 1, inplace=True)
```

```python
dfhappiness2020_2.rename(columns={'totalScore':'educationScore'},inplace=True)
```

```python
dfhappiness2020_2.head()
```

```python
dfhappiness2020_2['scaled_educationScore'] = (dfhappiness2020_2['educationScore']-dfhappiness2020_2['educationScore'].min())/(dfhappiness2020_2['educationScore'].max()-dfhappiness2020_2['educationScore'].min())
```

```python
dfhappiness2020_2.head()
```

```python
dfhappiness2020_2.plot(kind = 'barh', y = 'scaled_educationScore', figsize= (10,30))
```

```python
sns.heatmap(dfhappiness2020_2.corr())
```

```python
sns.lmplot(x= 'Ladder score', y='Perceptions of corruption', data = dfhappiness2020_2,order=2) #Polynomial regression plot
```
