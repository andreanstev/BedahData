Daily Value: DV, a term on food labels based on the RDA (Recommended Dietary Allowance) designed to help consumers use food label information to plan a healthy diet.

The Daily Value serves as a basis for declaring on the label the percent of the DV for each nutrient that a serving of the food provides. For example, the Daily Value for fat, based on a 2,000-calorie diet, is 65 grams (g). A food that has 13 g of fat per serving would state on the label that the "percent Daily Value" for fat is 20 percent.

https://www.medicinenet.com/script/main/art.asp?articlekey=14107

AKG, atau yang kadang juga ditulis Daily Values (DV), merujuk pada kebutuhan energi harian rata-rata yaitu sebesar 2000 kalori. AKG menunjukkan jumlah kadar nutrisi baik dalam satuan berat seperti milligram (mg) atau gram (gr), maupun penyajian dalam bentuk presentase (%) AKG.

Setiap nutrisi memiliki rekomendasi jumlah konsumsi harian masing-masing, sedangkan persentase AKG diperoleh berdasarkan proporsi kandungan nutrisi dalam sajian, dibandingkan rekomendasi jumlah konsumsi harian. Kebutuhan nutrisi harian dikatakan terpenuhi jika % AKG dari nutrisi tersebut mencapai 100%.

https://hellosehat.com/hidup-sehat/nutrisi/cara-membaca-label-informasi-nilai-gizi-pada-kemasan-makanan/

```python
from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
df = df[df['Category'] != 'Smoothies & Shakes']
df = df[df['Category'] != 'Coffee & Tea']
df = df[df['Category'] != 'Beverages']
df = df[df['Category'] != 'Desserts']
```

```python
# Drop columns that are not being used in analysis
df = df.drop(['Calories from Fat',
       'Total Fat', 'Total Fat (% Daily Value)', 'Saturated Fat',
       'Saturated Fat (% Daily Value)', 'Trans Fat', 'Cholesterol',
       'Cholesterol (% Daily Value)', 'Sodium', 'Sodium (% Daily Value)',
       'Carbohydrates', 'Carbohydrates (% Daily Value)', 'Dietary Fiber',
       'Dietary Fiber (% Daily Value)', 'Sugars', 'Protein',
       'Vitamin A (% Daily Value)', 'Vitamin C (% Daily Value)',
       'Calcium (% Daily Value)', 'Iron (% Daily Value)'], axis=1)
```

```python
# Remove some characters from strings so Serving Size only contains numerical value
df['Serving Size'] = df['Serving Size'].str.slice_replace(stop=7, repl='')
```

```python
# Remove some characters from strings so Serving Size only contains numerical value
df['Serving Size'] = df['Serving Size'].str.replace('(', '')
df['Serving Size'] = df['Serving Size'].str.replace(')', '')
df['Serving Size'] = df['Serving Size'].str.replace('g', '')
```

```python
df['Serving Size'] = pd.to_numeric(df['Serving Size'])
```

```python
df.corr()
```

```python
# See if Calories is a good predictor for Serving Size
import seaborn as sns
sns.regplot(x=df['Serving Size'], y=df['Calories'])
plt.ylim(0,)
```

```python
# Dataframe with only 2 columns
cdf = df[['Serving Size', 'Calories']]
```

```python
# Generate histogram chart for exploration
cdf.hist()
plt.show()
```

```python
# See how linear the correlation
plt.scatter(cdf['Serving Size'] ,cdf['Calories'],  color='blue')
plt.xlabel("Serving Size")
plt.ylabel("Calories")
plt.show()
```

```python
# Train test split
msk = np.random.rand(len(df)) < 0.8
train = cdf[msk]
test = cdf[~msk]
```

```python
# Train data distribution
plt.scatter( train['Serving Size'], train['Calories'], color='blue')
plt.xlabel("Serving Size")
plt.ylabel("Calories")
plt.show()
```

```python
#Linear regression modelling
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['Serving Size']])
train_y = np.asanyarray(train[['Calories']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
# Output
plt.scatter( train['Serving Size'], train['Calories'], color='blue')
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Serving Size")
plt.ylabel("Calories")
```

```python
# Evaluation with MSE
from sklearn.metrics import r2_score
test_x = np.asanyarray(test[['Serving Size']])
test_y = np.asanyarray(test[['Calories']])
test_y_hat = regr.predict(test_x)

print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_hat - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_hat - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_hat , test_y) )
```
